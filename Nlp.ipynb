```python
text1 = "Ethics are built right into ideals and objectives of the United Nations"
```

```python
len(text1)
```

```python
text2 = text1.split(' ')
len(text2)
```

```python
[w for w in text2 if len(w) > 3 ] # words that are greater than 3 letters long in text2
```

```python
[w for w in text2 if w.istitle()] # Capitalised words in text2
```

```python
[w for w in text2 if w.endswith('s')] # Words that ends with letter s
```
