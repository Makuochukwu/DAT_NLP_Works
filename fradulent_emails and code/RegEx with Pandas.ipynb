```python
for item in contents:
    emails_dict = {}
    sender = re.search(b'From:.*', item)
    #print(sender)
    if sender is not None:
        s_email = re.search(b'\w\S*@.*\w', sender.group())
        s_name = re.search(b'\".*\"', sender.group())
    else:
        s_email = None
        s_name = None
    if s_email is not None:
        sender_email = s_email.group()
    else:
        sender_email = None 
    emails_dict['sender_email'] = sender_email 
    print(emails_dict)
```
