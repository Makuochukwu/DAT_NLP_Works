# Filtering product reviews that are at least 500

```python
def circ(r):
    return 2*math.pi*r
```

```python
circ(50) #this is the value for the circumference to be passed.
```

```python
circl = lambda r: 2*math.pi*r    #another way, lambda here is an anonymous function, just incase you do not want to name your function.
```

```python
circl(50)
```

```python
multpl = lambda x,y: x*y  #finding the product of x and y
```

```python
multpl(10,10)
```

```python
rate.groupby('name')
```

```python
rate = rate.groupby('name').filter(lambda x: len(x)>500)
```

```python
rate.head() 
```

```python
rate['name'].unique()  # unique here is used to identify the individual product.
```

```python
len(rate['name'].unique())
```
