```python
import requests
from bs4 import BeautifulSoup
#from gensim.summarization import summarize
```

```python
url = 'https://www.bbc.com/news/world-us-canada-61588274'
```

```python
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page)
soup
```

```python
header = soup.find('h1').text
```

```python
p_tags = soup.find_all('p')
p_tags
```

```python
text = [tags.text.strip() for tags in p_tags]
text
```

```python
sentence_list = [sentence for sentence in text if not '\n' in sentence]
```

```python
sentence_list = [sentence for sentence in sentence_list if '.' in sentence]
```

```python
article = ''.join(sentence_list)
```

# Tokenization

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(article)
```

```python
spacy.explain?
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    print(token.text, '--->', token.pos_,'--->', spacy.explain(token.pos_))
```

# Named Entity

```python
for token in doc:
    print(token)
```

```python
doc.ents
```

```python
for ent in doc.ents:
    print(ent.text, '---->', ent.label_)
```

# Visualization

```python
from spacy import displacy
```

```python
displacy.render(doc, style='ent', jupyter=True)
```

```python
displacy.render(doc, style = 'dep', jupyter=True)
```

```python
lis = []
```

```python
for token in doc:
    data_dictionary = {}
    data_dictionary['Token'] = token
    data_dictionary['POS'] = token.pos_
    data_dictionary['Dependency'] = token.dep_
    data_dictionary['Explanation'] = spacy.explain(token.dep_)
    lis.append(data_dictionary)
lis
```
